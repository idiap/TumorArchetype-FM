```python
import pandas as pd
import anndata as ad

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

import matplotlib
matplotlib.rcParams["pdf.fonttype"] = 42
matplotlib.rcParams["ps.fonttype"] = 42
```

```python
## UMAP with patient A from UNI

emb = ad.read_h5ad("../results/supp/with_patient_A/uni/image_embedding.h5ad")
```

```python
palette = sns.color_palette("Accent", 7)

#insert deepskyblue at the beggining of the palette
palette.insert(0, 'deepskyblue')
sns.set_palette(palette)
palette
```

```python
sns.scatterplot(x=emb.obsm["umap"][:,0], y=emb.obsm["umap"][:,1], hue=emb.obs["tumor"], s=1, palette=palette)
sns.despine()
plt.title("UMAP from UNI embeddings with patient A", weight="bold")
plt.legend(title="Tumor", loc="upper left", bbox_to_anchor=(1,1))

plt.savefig("../Figures/Supp/umap_scatter_UNI_with_patient_A.pdf", bbox_inches="tight")
```

```python
sns.kdeplot(x=emb.obsm["umap"][:,0], y=emb.obsm["umap"][:,1], hue=emb.obs["tumor"], s=1, palette=palette)
sns.despine()
plt.title("UMAP from UNI embeddings with patient A", weight="bold")
sns.move_legend(plt.gca(), loc="upper left", bbox_to_anchor=(1, 1), title="Tumor")

plt.savefig("../Figures/Supp/umap_kde_UNI_with_patient_A.pdf", bbox_inches="tight")
```
